## Табличные данные. Библиотека Pandas

```python
import pandas as pd

%matplotlib inline
```

```python
# https://www.kaggle.com/c/titanic/data

data = pd.read_csv('./train.csv')
data.head(1)
```

```python
data
```

```python
print('Количество пассажиров: ', data.PassengerId.nunique())
print('Процент выживших: {} %'.format(int(100 * data['Survived'].sum() / data.PassengerId.nunique())))
```

```python
data['Age'].hist(figsize=(20, 10), edgecolor='white', grid=False, bins=80)
data[data['Sex'] == 'male']['Age'].hist(figsize=(20, 10), edgecolor='white', grid=False, bins=80, color='red')
```

```python
ax = data['Age'].hist(figsize=(20, 10), bins=80, edgecolor='white', grid=True)

data[data['Sex'] == 'female']['Age'].hist(figsize=(20, 10), bins=80, 
                                          edgecolor='white', grid=True, color='red')
ax.legend(['Male', 'Female'])
```

```python
ax = data['Age'].hist(figsize=(20, 10), bins=80, edgecolor='white', grid=True)

data[data['Survived'] == 0]['Age'].hist(figsize=(20, 10), bins=80, 
                                          edgecolor='white', grid=True, color='green')
ax.legend(['Survived', 'Not Survived'])
```

```python
data.Pclass.hist(figsize=(10, 5), grid=False)
data[data['Survived'] == 0].Pclass.hist(color='red', grid=False)
```

```python
series = round(100 * data.groupby(['Pclass'])['Survived'].sum() / data.groupby(['Pclass'])['PassengerId'].nunique())
series
```

```python
pd.DataFrame(series)
```

```python
series = data.groupby(['Cabin'])['Survived'].sum() / data.groupby(['Cabin'])['PassengerId'].nunique()
```

```python
data.Cabin = data.Cabin.fillna('N/A')
```

```python
data[data.Pclass == 3].groupby('Cabin')['Cabin'].count()
```

```python
series.hist()
```

```python
ax = data[data.Pclass == 3]['Age'].hist(figsize=(20, 10), bins=80, edgecolor='white', alpha=0.5)
data[data.Pclass == 2]['Age'].hist(figsize=(20, 10), bins=80, edgecolor='white', alpha=0.5)
data[data.Pclass == 1]['Age'].hist(figsize=(20, 10), bins=80, edgecolor='white', alpha=0.5)

ax.legend(['3', '2', '1'])
```

```python
data.groupby(['Pclass'])['Age'].hist(figsize=(10, 5), alpha=0.5, bins=80)
```

```python
data.groupby(['Pclass'])['Age'].median()
```

```python
data['Family'] = data['SibSp'] + data['Parch']
```

```python
data[data.Pclass == 3]['Family'].hist(figsize=(10, 5), bins=100)
data[(data.Pclass == 3) & (data.Survived == 0)]['Family'].hist(figsize=(10, 5), bins=100)
```

```python
series = data.groupby(['Family'])['Survived'].sum() / data.groupby(['Family'])['Survived'].count()
```
